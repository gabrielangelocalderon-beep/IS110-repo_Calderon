```python
total = 0
count = 0

while True:
    user_input = input("Enter a number: ")

    if user_input == "done":
        break

    try:
        num = float(user_input)
        total = total + num
        count = count + 1
    except:
        print("Invalid input")
        continue

average = total / count
print(total, count, average)
```

```python
numbers = []

while True:
    user_input = input("Enter a number: ")

    if user_input == "done":
        break

    try:
        num = float(user_input)
        numbers.append(num)
    except:
        print("Invalid input")
        continue

if numbers:  # check if the list is not empty
    print("Maximum:", max(numbers))
    print("Minimum:", min(numbers))
else:
    print("No valid numbers were entered.")
```
