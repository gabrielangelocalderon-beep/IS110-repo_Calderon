```python
def compute_pay(hours, rate):
    if hours > 40:
        overtime = hours - 40
        pay = (40 * rate) + (overtime * rate * 1.5)
    else:
        pay = hours * rate
    return pay

hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

pay = compute_pay(hours, rate)
print("Pay:", pay)
```

```python
def computegrade(score):
    if score >= 97 and score <= 100:
        return "1.0"
    elif score >= 93 and score < 97:
        return "1.25"
    elif score >= 89 and score < 93:
        return "1.5"
    elif score >= 85 and score < 89:
        return "1.75"
    elif score >= 81 and score < 85:
        return "2.0"
    elif score >= 77 and score < 81:
        return "2.25"
    elif score >= 74 and score < 77:
        return "2.5"
    elif score >= 0 and score < 74:
        return "3.0"
    else:
        return "Invalid score"

while True:
    try:
        score = float(input("Enter Score: "))
        print("Grade:", computegrade(score))
    except:
        print("Invalid input")
```
