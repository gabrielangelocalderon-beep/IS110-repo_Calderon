```python
x = int(input("Enter a number: "))

if x > 50:
    print('Above 50')
elif x > 20:
    print('Above 20')
elif x > 10:
    print('Above 10')
elif x > 0:
    print('Above 0')
else:
    print('Something else')
```

```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
    overtime_hours = hours - 40
    overtime_pay = overtime_hours * rate * 1.5
    regular_pay = 40 * rate
    pay = regular_pay + overtime_pay
else:
    pay = hours * rate

print("Pay:", pay)
```

```python
score = float(input("Enter score (0-100): "))

if score >= 98 and score <= 100:
    print("Grade: 1.0")
elif score >= 94 and score < 98:
    print("Grade: 1.2")
elif score >= 90 and score < 94:
    print("Grade: 1.5")
elif score >= 86 and score < 90:
    print("Grade: 1.75")
elif score >= 82 and score < 86:
    print("Grade: 2.0")
elif score >= 78 and score < 82:
    print("Grade: 2.25")
elif score >= 75 and score < 78:
    print("Grade: 3.0")
elif score >= 0 and score < 75:
    print("Grade: 5.0")
else:
    print("Invalid score")
```
